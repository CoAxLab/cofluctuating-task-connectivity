```python
import numpy as np
import matplotlib.pylab as plt
from nilearn import image
from scipy.spatial.distance import euclidean, correlation, squareform
```

```python
final_subjects = np.loadtxt("../data/subjects_intersect_motion_035.txt")
```

```python
# Load intercepts for each case, the hypothesis is that these should be closer to 

# STROOP
pattern = "../task-stroop/first-level-nilearn/edge_intercept/sub-%d/constant/cope.nii.gz"
stroop_before = [squareform(np.squeeze(image.load_img(pattern % subj).get_fdata()), 
                               checks=False) for subj in final_subjects]

pattern = "../task-stroop/first-level-nilearn/edge/sub-%d/constant/cope.nii.gz"
stroop_after = [squareform(np.squeeze(image.load_img(pattern % subj).get_fdata()), 
                               checks=False) for subj in final_subjects]

# MSIT
pattern = "../task-msit/first-level-nilearn/edge_intercept/sub-%d/constant/cope.nii.gz"
msit_before = [squareform(np.squeeze(image.load_img(pattern % subj).get_fdata()), 
                               checks=False) for subj in final_subjects]

pattern = "../task-msit/first-level-nilearn/edge/sub-%d/constant/cope.nii.gz"
msit_after = [squareform(np.squeeze(image.load_img(pattern % subj).get_fdata()), 
                               checks=False) for subj in final_subjects]

# RESTING
pattern = "../task-rest/first-level-nilearn/edge/sub-%d/constant/cope.nii.gz"
intrinsic_resting = [squareform(np.squeeze(image.load_img(pattern % subj).get_fdata()), 
                               checks=False) for subj in final_subjects]
```

```python
plt.hist([correlation(a,b) for (a,b) in zip(stroop_before, intrinsic_resting)], label="before")
plt.hist([correlation(a,b) for (a,b) in zip(stroop_after, intrinsic_resting)], label="after")
plt.legend()
```
